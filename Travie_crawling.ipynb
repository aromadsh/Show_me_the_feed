```python
import re
import requests
from bs4 import BeautifulSoup

# Travie Essay 탭 접근해서 각 기사에 대한 하이퍼링크 따기
hdr={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/63.0.3239.132 Safari/537.36'}
url = 'https://www.travie.com/news/articleList.html?page=2&total=415&box_idxno=&sc_sub_section_code=S2N12&view_type=sm'

response = requests.get(url, headers=hdr)
val = BeautifulSoup(response.text, 'html.parser')

# 각 기사글에 대한 URI를 추출하여 URL에 붙여넣어서 각 기사에 접근
article_list = val.find_all('h4')
href_list = []
text_list = []
for article in article_list:
    href_list.append(article.find('a')['href'])
    

for cnt, href in enumerate(href_list):
    article_url = f'https://www.travie.com{href}'
    response = requests.get(article_url, headers=hdr)
    val = BeautifulSoup(response.text, 'html.parser')
    
    # 각 기사의 링크에 내용이 담긴 부분의 태그에만 접근하여 전처리를 거친 후 텍스트만 저장하는 리스트에 추가
    p_tag = val.find_all('p')
    for tag in p_tag[:-3]:
        if tag.find('strong') != None:
            continue
        text=re.sub(r'[\r\n\xa0]',' ',tag.text)
        text=re.sub(r'[^0-9 가-힣.,\'\'""]','',text)
        text_list.append(text)
```

```python
print(text_list)
```

```python
# 텍스트파일을 저장용으로 생성한 후 저장
file = open('page.txt', 'w', encoding='utf-8')
for line in text_list:
    file.write(line)

file.close()
```
